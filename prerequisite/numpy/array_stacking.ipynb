# Array Stacking

```python
import numpy as np
a1 = np.arange(0,25).reshape(5,5)
a2 = np.linspace(3,4, 10).reshape(2,5)
a3 = np.linspace(2,3, 25).reshape(5,5)
a5 = np.linspace(3,4, 15).reshape(5,3)
a1, a2, a3, a5
```

### Vartical Stack

```python
aV = np.vstack((a1, a2, a3))
aV
```

```python
aV = np.vstack((a1, a2, a3, a5)) # ERROR (Vertically axises must match)
aV
```

```python
b1, b2, b3 = np.vsplit(aV,3) # All elements will get equally devided
b1, b2, b3
```

### Horizontal Stack

same as vstack

```python
aH = np.hstack((a1, a3, a5)) # (Horizontally axises must match)
aH
```

### Concatenate

```python
cV = np.concatenate((a1, a2, a3), axis=0) # Same as vstack()
cH = np.concatenate((a1, a5, a3), axis = 1) # Same as hstack()
cV.shape, cH.shape, cV, cH
```

### Stack

stack() function stacks arrays in a new sxis (all array have to have the same shape)

```python
dS = np.stack((a1, a3))
dS.shape, dS
```

### Transposing

```python
print(dS.shape)
dS = dS.transpose(2,0,1)
print(dS.shape)
```

```python
dS = dS.transpose(0, 2, 1)
print(dS.shape)
dS = dS.transpose()
print(dS.shape)
```

```python
print(dS.shape)
dS = dS.swapaxes(0, 1)
print(dS.shape)
```
