```python
normalDistribution = np.random.randn(100*100000)
uniformDistribution = np.random.rand(1000*100000)
print(normalDistribution)
print(uniformDistribution)
```

```python
print("Mean of the normalDistribution: "+str(normalDistribution.mean()))
print("Mean of the uniformDistribution: "+str(uniformDistribution.mean()))
print("Theorytical mean of a normal distribution: 0")
print("Theorytical mean of a uniform distribution: "+str((uniformDistribution.min() + uniformDistribution.max()) / 2))
```

```python
plt.hist(normalDistribution, density= True, bins=400, histtype="step", color="blue", label="NormalDistribution")
plt.hist(uniformDistribution, bins=100, density=True, color="red", histtype="step", label="UniformDistribution")
plt.legend(loc = "upper left")
```

### Mathematical and statistical functions

```python
for func in (normalDistribution.min, normalDistribution.max, normalDistribution.sum, normalDistribution.prod, normalDistribution.std, normalDistribution.var, normalDistribution.mean):
    print(func.__name__, ": ", func())
```

```python
for func in (uniformDistribution.min, uniformDistribution.max, uniformDistribution.sum, uniformDistribution.prod, uniformDistribution.std, uniformDistribution.var, uniformDistribution.mean):
    print(func.__name__, ": ", func())
```
