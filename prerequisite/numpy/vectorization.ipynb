# Vectorization

Instead of one by one array multiplication we can use vectorization method to compute parallelly, it makes code faster.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
a = np.empty((4080, 1440))
a
```

```python
for i in range(4080):
    for j in range(1440):
        a[i, j] = math.sin(i * j / (40.5)) # very inefficient (Bad example)

a
```

```python
xC = np.arange(4080)
yC = np.arange(1440)
x, y = np.meshgrid(xC, yC)
x, y
```

```python
data = np.sin(x, y / 40.5) # The correct way to do.
data
```

```python
xC = np.arange(1024)
yC = np.arange(768)
x, y = np.meshgrid(xC, yC)
data = np.sin(x, y / 40.5)
data
```

```python
fig = plt.figure(1, figsize=(7, 6))
plt.imshow(data, cmap="hot")
plt.show()
```
