# Iteration

```python
import numpy as np
x = np.linspace(3,678, 280).reshape(4,7,10)
x
```

```python
i = 1
for firstAxis in x: # This will iterate in respect of ist axis/dimension has 4 elements 
    print(i, ": ", firstAxis)
    i += 1
```

```python
i=1
for item in x.flat: # access individual number of an ndarray
    print(i, ": ", item)
    i = i+1
```
