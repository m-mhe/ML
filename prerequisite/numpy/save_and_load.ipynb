# Save & Load

```python
import numpy as np
a = np.random.randn(4840, 2160)
b = np.random.rand(4840, 2160)
a, b
```

```python
np.save("array_a", a)
x = np.load("array_a.npy")
x
```

```python
np.savetxt("arrayB.csv", b)
x = np.loadtxt("arrayB.csv") # This method is slower than binary format.
x
```

```python
np.savez("array_a_b.npz", arrayA = a, arrayB = b)
```

```python
x = np.load("array_a_b.npz")
x["arrayA"], x["arrayB"]
```
