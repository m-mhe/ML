# Array Indexing

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.linspace(2, 299, 25).reshape(5,5)
a1D = a.ravel()
plt.plot(a1D)
print(a1D)
a
```

### Indexing Through a 1D Array

```python
a1D[21] # Single element array[indexNumber]
```

```python
a1D[14:21] # To get multiple elements (slice of an array) array[startIndex(include): endIndex(exclude)]
```

```python
a1D[:22] # A slice from index 0 to index 21
```

```python
a1D[14:] # A slice from index 14 to the last index
```

```python
print(a1D[-1]) # Last index
a1D[-2] # Second last index
```

```python
b = a1D[20:-1]
b # Pointing to the same data (a1D)
```

```python
b = a1D[15:-1].copy()
b # Not pointing to the same data (a1D)
```

### Indexing Through a Multidimantional Array

Every things is almost the same as 1D array, just we need to do indexing for each dimantion (axis).

```python
print(a[2,3])
a[4,4] # Indexing a single element
```

```python
print(a[:3,-3:])
a[2:5,4] # Indexing a slice of elements
```

```python
print(a[(3,1),(2,3)]) # In this case elements are from index a[3,2] and index a[1,3]
print(a[3,2],"&", a[1,3])
a[(2,4),4] # This is called fancy indexing (here indexing happens in pair, in this case elements are from index a[2,4] and index a[4,4])
```

```python
a[[False, True, True, False, True],:] # This is called boolean indexing
```

```python
a[[False, True, True, False, True],[False, True, True, False, True]]
```

```python
row = np.array([False, True, True, False, True])
col= np.array([True]*5)
a[row,col] # Boolean indexing doesn't work on multiple axis (hence ERROR)
```

```python
a[np.ix_(row, col)]
```
