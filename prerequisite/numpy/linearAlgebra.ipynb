```python
import numpy as np
import numpy.linalg as linalg
```

# Linear Algebra

```python
x = np.arange(45).reshape(5, 9)
x
```

```python
x = x.T # same as transpose, but only works on upto 2D arrays
x
```

### Matrix Multiplication

```python
y = np.arange(10).reshape(5,2)
y
```

```python
z = x.dot(y)
```

### Determinant

```python
z = np.linspace(1, 278, 625).reshape(25, 25)
print(z)
x = linalg.det(z)
x
```
