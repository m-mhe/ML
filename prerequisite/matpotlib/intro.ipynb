# Introduction

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2, 24, 10740)
plt.plot(x) # As simple as poloting with just array.
x
```

```python
y = np.sin(x)
plt.plot(x, y) # Now here this .plot() function took first array on x axis and y second array on axis. If we do not mention two arrays (like previous one) for a Line2D graph then that one array would be taken in y axis and continous number would be in x axis by default.
y
```

### Configuring .axis() function to manipulate number range of x & y axis

```python
plt.plot(x, y)
plt.axis([1, 25, -5, 5])
plt.title("Sine Function")
plt.xlabel("x")
plt.ylabel("y = sin(x)")
plt.legend("y")
plt.grid(True)
```

### Many in one

```python
x = np.linspace(-1.4, 1.4, 25)
line1, line2, line3 = plt.plot(x, x, 'g-', x, x**2, 'r--', x, x**3, 'b^')
line3.set_alpha(0.2)
line2.set_dash_capstyle("round")
plt.savefig("plot_number_one_expo.png")
```

### SubPlots

```python
x = np.linspace(2, 24, 10740)
plt.figure(figsize=(15,15))
plt.subplot(3, 3, 1)
plt.plot(x, np.sin(x))
plt.title("sine(x)")
plt.axis([2, 24, -4, 4])

plt.subplot(332)
plt.plot(x, np.cos(x))
plt.title("cosine(x)")
plt.axis([2, 24, -4, 4])

plt.subplot(133)
plt.plot(x, np.tan(x))
plt.title("tan(x)")
plt.axis([2, 12, -5, 5])

plt.subplot(334)
y = 1/(np.tan(x))
plt.plot(x, y)
plt.title("cot(x)")
plt.axis([2, 15, -15, 15])


plt.subplot(335)
y = 1/(np.cos(x))
plt.plot(x, y)
plt.title("sec(x)")
plt.axis([2, 15, -15, 15])

plt.subplot(337)
y = 1/(np.sin(x))
plt.plot(x, y)
plt.title("cosec(x)")
plt.axis([2, 15, -15, 15])

x = np.linspace(-24, 24, 5684)
plt.subplot(338)
plt.title("x^2")
plt.plot(x, x**2)
```

### Multiple figures

```python
plt.figure(1, figsize= (3,3))
plt.plot(x, x**2)
plt.grid(True)

plt.figure(2, figsize= (3,3))
plt.plot(x, x**3)
plt.grid(True)
plt.savefig("two_to_the_power_three.png")

plt.figure(3, figsize= (3,3))
plt.plot(x, x**4)
plt.grid(True)

plt.figure(4, figsize= (3,3))
plt.plot(x, x**5)
plt.grid(True)

plt.figure(5, figsize= (3,3))
plt.plot(x, x**6)
plt.grid(True)

plt.figure(6, figsize= (3,3))
plt.plot(x, x**1)
plt.grid(True)
```
