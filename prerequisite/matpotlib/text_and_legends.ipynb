# Text & Legend

We can add text at any point of a graph by twiking .text() function

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-25,25, 1000000)
y = x**3 #np.sin(x)
plt.plot(x, y)
plt.axis([-3.5, 3.5, -15, 15])
plt.title("Cube Function")
plt.text(2.5, 2, "x = 2.5", weight = "heavy")
plt.text(2.5, 0, "y = 0", weight = "heavy")
```

After giving each line2D a label we can show legends just by calling .legend() function

```python
np.random.seed(67)
x= np.random.randn(92451995)
y = np.random.rand(92455)
plt.figure(0, figsize=(25, 25))
plt.title("Distributions")
plt.hist(x, histtype= "step", density=True, bins=600, label= "Normal Distribution")
plt.hist(y, histtype="step", bins=100, label="Unifide Distribution", density= True)
plt.legend(loc = "best")
plt.savefig("distribution.png")
```
