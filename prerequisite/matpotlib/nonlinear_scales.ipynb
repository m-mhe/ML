# Nonlinear Scales

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,15, 500)
y = x*3/np.exp(2*x)

plt.figure(0, figsize=(10, 10))

plt.subplot(1, 2, 1)
plt.plot(x, y )
plt.title("log")
plt.yscale("log")

plt.subplot(2, 2, 2)
plt.plot(x, y )
plt.title("logit")
plt.yscale("logit")

plt.subplot(2, 2, 4)
plt.plot(x, y )
plt.title("symlog")
plt.yscale("symlog", linthresh = 0.05)

plt.savefig("logarithm.png")
```
